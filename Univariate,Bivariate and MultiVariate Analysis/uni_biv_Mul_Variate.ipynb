# Important Libraries 

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.drop(df.columns[0],axis="columns")
```

```python
df
```

```python
df.columns = ['sepal_length','sepal_width','petal_length','petal_width','species']
```

```python
df
```

## Univariate Analysis 

```python
df_iris_setosa = df.loc[df['species'] == 'Iris-setosa']
df_iris_setosa
```

```python
df_iris_virginica = df.loc[df['species'] == 'Iris-virginica']
df_iris_virginica 
```

```python

df_iris_versicolor = df.loc[df['species'] == 'Iris-versicolor']
df_iris_versicolor
```

```python
plt.plot(df_iris_setosa['sepal_length'],np.zeros_like(df_iris_setosa['sepal_length']),'o')
plt.plot(df_iris_versicolor['sepal_length'],np.zeros_like(df_iris_setosa['sepal_length']),'o')
plt.plot(df_iris_virginica['sepal_length'],np.zeros_like(df_iris_setosa['sepal_length']),'o')
plt.xlabel("Sepal length")
plt.show()


```

## Bivariate Analysis

```python

sns.FacetGrid(df, hue="species", height=5)\
.map(plt.scatter, "petal_length", "sepal_width")\
.add_legend()

plt.show()
```

### Multivariate Analysis 

```python
mul = sns.pairplot(df, hue="species",height=3)
mul
# sns.pairplot() -->Creates a matrix of plots comparing every numeric feature
```
